```python
#converting my upper case into lower case chr and print the full name
def upper_to_lower(s):
  ans = ""
  for i in s:
    if ord(i)>=65 and ord(i)<=90:
      lw_asc = ord(i)+32
      ans += chr(lw_asc)
    else:
      ans += i
  print(ans)
```

```python
#converting my upper case into lower case chr and lower to upper case
def upper_to_lower_case(s):
  ans = ""
  for i in s:
    if ord(i)>=65 and ord(i)<=90:
      lw_asc = ord(i)+32
      ans += chr(lw_asc)
    elif ord(i)>=97 and ord(i)<=122:
      up_asc = ord(i)-32
      ans += chr(up_asc)
  print(ans)
```

```python
upper_to_lower_case('sUJItha')
```

```python
upper_to_lower_case('HaRI KrisHNAN')
```

```python
class Eludi:
  def __init__(self, p1, p2):
    self.point1 = p1
    self.point2 = p2
  def calculate(self):
    x1,y1 = self.point1
    x2,y2 = self.point2

    distance=  ((x2-x1)**2 + (y2 - y1)**2)**0.5

    return distance

```

```python
class rectangle:
  def __init__(self,h, w):
    self.height = h
    self.width = w
  def calculate(self):
    x1, y1 = self.height
    x2, y2 = self.width

    X_center = x1+x2 /2.0
    y_center = y1+y2 / 2.0

    return (X_center, y_center)
```

```python
height_size = (1,4)
width_size = (5,6)
answer = rectangle(height_size,width_size)
center_point = answer.calculate()
print(center_point)
```
